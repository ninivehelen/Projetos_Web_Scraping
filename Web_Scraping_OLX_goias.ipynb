**Web Scraping. Anúncios de lotes em Goías**
*</br>A coleta de dados web, ou raspagem web, é uma forma de mineração que permite </br> a extração de dados de sites da web convertendo-os</br> em informação estruturada para posterior análise.*

```python
#Importando bibliotecas 
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
from urllib.error import URLError, HTTPError
import pandas as pd
import requests
import json
```

```python
#Lista para guardar os dados
olx_dados = []
#Quantidade de paginas que serão coletadas os dados
quantidade_paginas = 100

#For para pecorrer as paginas
for i in range(quantidade_paginas):
    #Url da pagina principal 
    url = 'https://go.olx.com.br/?o='+  str(i + 1) + '&q=lote'
    #Cabeçalho para a solicitação dos dados
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}

    #Vefiricando a conexao
    try:
        req = Request(url, headers= headers)
        response = urlopen(req)
        html = response.read()
        
    except HTTPError as e:
        print(e.status, e.reason)

    except URLerror as e:
        print(e.reason)
    #Guardado acesso a pagina
    soup = BeautifulSoup(html, 'lxml')

    #For para coletar os links para acessar os anúncios dos lotes
    for link in soup.findAll('li', attrs={'class': 'sc-1fcmfeb-2 iezWpY'}):
      for a in link.findAll("a"):
        href = a.attrs["href"]
        #print(href)
        
        headers2 = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}
        #Cabeçalho para solicitação nos links coletados.

    #Verificando a conexão
    try:
        req2 = Request(href, headers= headers)
        response = urlopen(req2)
        html2 = response.read()
        
    except HTTPError as e:
        print(e.status, e.reason)

    except URLerror as e:
        print(e.reason)

    #Guardado acesso aos links
    soup2 = BeautifulSoup(html2, 'lxml')
    #For para coletar os titulos 
    for link2 in soup2.findAll('div', attrs={'class': 'sc-18p038x-3 dSrKbb'}):
       #Coletando os titulos
        titulo = link2.find('h1', attrs={'class': "sc-45jt43-0 eCghYu sc-ifAKCX cmFKIN"})
        #Coletando o preço
        preco =  link2.find('h2', attrs={'class': "sc-ifAKCX eQLrcK"})
        if preco is not None:
          preco2 = preco
        #print("Nome\n",nome.text)
        #print("Preço\n",preco.text)
    #For para coletar a descrição 
    for link3 in soup2.findAll('p', attrs={'class': 'sc-1kv8vxj-0 gDWTkJ'}):
      #Coletando a descrição
      descricao =  link3.find('span', attrs={'class': "sc-1sj3nln-1 eOSweo sc-ifAKCX cmFKIN"})
      print("Descrição\n",descricao.text)
   #Coletando localizacao
    loca = soup2.find('div', attrs={'class': 'sc-hmzhuo gqoVfS sc-jTzLTM iwtnNi'})
    #For para coletar a localização
    for loca2 in loca.findAll('div', attrs={'data-testid': 'ad-properties'}):
      for loca3 in loca2.findAll('div', attrs={'class': 'duvuxf-0 h3us20-0 jyICCp'}):
       for loca4 in loca3.findAll('div', attrs={'class': 'sc-hmzhuo sc-1f2ug0x-3 ONRJp sc-jTzLTM iwtnNi'}):
         #Coletando localização
        print("\n",loca4.text)
    #Coletando informçao de tamanho dos lotes  
    tam = soup2.find('div', attrs={'class': 'h3us20-2 fMOiyI'})
    for tam2 in tam.findAll('div', attrs={'data-testid': 'ad-properties'}):
     for tam3 in tam2.findAll('div', attrs={'class': 'duvuxf-0 h3us20-0 jyICCp'}):
       for tam4 in tam3.findAll('div', attrs={'class': 'sc-hmzhuo sc-1f2ug0x-3 ONRJp sc-jTzLTM iwtnNi'}):
         #Coletando informaçoes de lotes
        print("\n",tam4.text)
     #Adicionando informaçoes coletadas na lista 
     olx_dados.append({
                    "Titulo": titulo.text,
                    "Preço": preco2.text,
                    "Descricão:": descricao.text,
                    "Tamanho": loca4.text,
                    "Localização":tam4.text,
                    
                })
    #Imprimindo a lista
    
    #Salvando os dados coletados em um formado json
    with open('olx_dados.json', 'w', encoding='utf-8') as fp:
            json.dump(olx_dados, fp, indent=3, ensure_ascii=False)
        
          
    
    


    
    
    
     
        

        
            
      
     
```
